### 4: PERFORM DATA CLEANING

```python
#Let's drop the date
data = data.drop(['date', 'rating', 'length'],axis=1)
```

```python
data.head(20)
```

```python
variation_dummies = pd.get_dummies(data['variation'], drop_first = True)
# Avoid Dummy Variable trap which occurs when one variable can be predicted from the other. converting the categories by one hot encoding
```

```python
variation_dummies
```

```python
# first let's drop the column
data.drop(['variation'], axis=1, inplace=True) # so once one hot encoding is done variation table is drop
```

```python
# Now let's add the encoded column again
data = pd.concat([data, variation_dummies], axis=1)# concat the dummies columns into reviews df
```

```python
data
```

```python
# remove PUNCTUATION FROM TEXT, stopwords and applyinh count vectorization to the data
```

```python
import string
string.punctuation
```

```python
import nltk # Natural Language tool kit 

nltk.download('stopwords')
```

```python
# we have to download stopwords Package to execute this command
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
# Let's define a pipeline to clean up all the messages 
# The pipeline performs the following: (1) remove punctuation, (2) remove stopwords

def message_cleaning(message):
    Test_punc_removed = [char for char in message if char not in string.punctuation]
    Test_punc_removed_join = ''.join(Test_punc_removed)
    Test_punc_removed_join_clean = [word for word in Test_punc_removed_join.split() if word.lower() not in stopwords.words('english')]
    return Test_punc_removed_join_clean
```

```python

# applying the above function for the variation column
# Let's test the newly added function
data_clean = data['verified_reviews'].apply(message_cleaning)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# Define the cleaning pipeline we defined earlier
vectorizer = CountVectorizer(analyzer = message_cleaning)
reviews_countvectorizer = vectorizer.fit_transform(data['verified_reviews'])
```

```python
#first let's drop the column
data.drop(['verified_reviews'], axis=1, inplace=True)
reviews = pd.DataFrame(reviews_countvectorizer.toarray())
```

```python

# Let's drop the target label coloumns
X = data.drop(['feedback'],axis=1)
```

```python

X # getting my x and y columns for training and test data set
```

```python
y = data['feedback']
```

```python
# 10: TRAIN A NAIVE BAYES CLASSIFIER MODEL
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB

NB_classifier = MultinomialNB()
NB_classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_predict_train = NB_classifier.predict(X_train)
y_predict_train
cm = confusion_matrix(y_train, y_predict_train)
sns.heatmap(cm, annot=True)
```

```python
# Predicting the Test set results
y_predict_test = NB_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict_test))
```
